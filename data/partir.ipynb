```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset_final.csv')
```

```python
# # 2. Dividir 70% para Treino (train) e 30% para Teste (test)
# # O stratify=df['source_name'] garante que as percentagens das classes ficam iguais nos dois ficheiros!
# df_treino, df_teste = train_test_split(df, test_size=0.3, random_state=42, stratify=df['source_name'])
```

```python
# 1. Primeira Divisão: 70% para Treino e 30% para o conjunto Temporário (Validação + Teste)
# O stratify garante que as percentagens das classes ficam iguais
df_treino, df_temp = train_test_split(df, test_size=0.3, random_state=42, stratify=df['source_name'])
```

```python
# 2. Segunda Divisão: Cortar os 30% (df_temp) a meio -> 15% Validação e 15% Teste
df_val, df_teste = train_test_split(df_temp, test_size=0.5, random_state=42, stratify=df_temp['source_name'])
```

```python
# 3. Guardar os novos ficheiros
df_treino.to_csv('dataset_treino.csv', index=False)
df_val.to_csv('dataset_validacao.csv', index=False)
df_teste.to_csv('dataset_teste.csv', index=False)
```

```python
print(f"Tamanho do Treino (70%): {len(df_treino)} textos")
print(f"Tamanho da Validação (15%): {len(df_val)} textos")
print(f"Tamanho do Teste (15%): {len(df_teste)} textos")

print("\n--- Proporção das classes no ficheiro de VALIDAÇÃO ---")
print(df_val['source_name'].value_counts(normalize=True) * 100)

print("\n--- Proporção das classes no ficheiro de TESTE ---")
print(df_teste['source_name'].value_counts(normalize=True) * 100)
```
